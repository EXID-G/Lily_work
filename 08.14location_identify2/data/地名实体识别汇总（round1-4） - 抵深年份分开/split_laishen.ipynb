```python
df_info = pd.read_csv('../../raw_data/interviewee_infomation/roundAll_info_final.csv')
df_info.head(2)
```

```python
script_withmorePeople_noAge = df_info["转录稿文件名"].value_counts()[df_info["转录稿文件名"].value_counts()>1].index.tolist()

script_withmorePeople_noAge.extend(['3_SZTUWWX_邓先生王女士','3_SZTUWWX_谭女士霍先生','10_SZTU-lx_凹凸曼&李先生',"WLL_黄师傅-已校对（彭楚玥）","WZY_04-江","XRH_01_WangW"])
script_withmorePeople_noAge
```

```python
df_info["year"] = df_info["抵深年份"]
s_index = df_info[df_info["year"].str.endswith("s")].index
df_info.loc[s_index,"year"] = df_info[df_info["year"].str.endswith("s")]["year"].str[:-1]
```

```python
df_info.loc[~df_info["转录稿文件名"].isin(script_withmorePeople_noAge),["转录稿文件名","year"]].sort_values("year")
```

```python
my_dir = [ 'round1_csv',
 'round2_csv',
 'round3_csv',
 'round4_csv',
 ]

## mydir用完后已经被删除了

for dir in my_dir:
    for file in os.listdir(f"../地名实体识别汇总（round1-4）/{dir}"):
        file_name, file_extension = os.path.splitext(file)

        source_file_path = os.path.join(f"../地名实体识别汇总（round1-4）/{dir}", file)

        if file_name in script_withmorePeople_noAge:
            destination_file_path = os.path.join("others",file)
            shutil.copy(source_file_path, destination_file_path)
            continue

        # print(file_name)
        year = int(df_info[df_info["转录稿文件名"]==file_name]["year"].values[0])
        # print(year)

        if year >= 1970 and year < 1980:
            destination_file_path = os.path.join("1970-1980",file)
        if year >= 1980 and year < 1990:
            destination_file_path = os.path.join("1980-1990",file)
        if year >= 1990 and year < 2000:
            destination_file_path = os.path.join("1990-2000",file)
        if year >= 2000 and year < 2010:
            destination_file_path = os.path.join("2000-2010",file)
        if year >= 2010:
            destination_file_path = os.path.join("2010+",file)


        shutil.copy(source_file_path, destination_file_path)
```
